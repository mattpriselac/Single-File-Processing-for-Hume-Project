```python
import csv
```

```python
relative_total_outfile = open('RelativeTotalCites.csv', 'w')
raw_inFile = open('RawTotalCites.csv', 'r')
raw_reader = csv.reader(raw_inFile)
total_cites = 0
relative_score_list = []
for row in raw_reader:
    total_cites += float(row[1])
raw_inFile.close
print(total_cites)
```

```python
raw_inFile = open('RawTotalCites.csv', 'r')
raw_reader = csv.reader(raw_inFile)
for row in raw_reader:
    para = row[0]
    raw_score = round(float(row[1]), 5)
    relative_score = round(raw_score/total_cites, 5) * 100
    relative_score_list.append((para, relative_score))
raw_inFile.close()
relative_writer = csv.writer(relative_total_outfile)

for pair in relative_score_list:
    relative_writer.writerow(pair)
relative_total_outfile.close()
print('done!')
```
