```python
def parensCleaner(cite_in):
    num_check = re.compile('(\d{1,3})+([-–—,](\d{1,3}))*')
    p_cite_check = re.compile('(p{0,2}\.)*\d{1,3}')
    only_num = get_num = re.compile('\d+')
    #first, get cites that only have 'T'
    if cite_in.rawCitationText[1] == "T":
        #now make sure we only have a after the T
        if num_check.match(cite_in.rawCitationText) != None:
            pageNum = num_check.match(cite_in.rawCitationText).group()
            cite_in.cleanedCitation = pageNum

    #now get cites that start with a p or pp. or just a number
    elif p_cite_check.search(cite_in.rawCitationText) != None:
        #get rid of the single digits that are often part of laying out structure of an argument
        if len(p_cite_check.search(cite_in.rawCitationText).group()) > 1:
            #get the numbers only from the results of the page number group of the intiial re search
            num_only = only_num.search(p_cite_check.search(cite_in.rawCitationText).group()).group()
            cite_in.cleanedCitation = num_only

    
    
```

```python
total_norton_cites = 0
total_sbn_cites = 0
total_clean_parens = 0

for file in os.listdir('txts'):
    if '.txt' in file:
        obj = Paper('txts/'+file)
        obj.NortonSearch()
        total_norton_cites += len(obj.nortonCites)
        obj.SbnSearch()
        total_sbn_cites += len(obj.sbnCites)
        obj.parenthesesCapture()
        for cite in obj.rawParenthesesCapture:
            parensCleaner(cite)
            if cite.cleanedCitation != '':
                total_clean_parens += 1

print(total_norton_cites)
print(total_sbn_cites)
print(total_clean_parens)
```
